```python
n = input()

predict = [0 for i in range(500001)]
check_once = [0 for i in range(500001)]
estimate=[0 for i in range(500001)]

arr1 = list()
arr2 = list()

for i in range(1,int(n)+1):
    predict[i] = int(input())
    if(check_once[predict[i]]==0):
        check_once[predict[i]] = 1
        estimate[i] = predict[i]

for i in range(1,int(n)+1):
    if(check_once[i]==0):
        arr1.append(i)
    if(estimate[i]==0):
        arr2.append(i)

arr1.sort()
arr2.sort()

result =0

for i in range(0,len(arr1)):
    result += abs(arr1[i]-arr2[i])
    
print(result)
print(predict)
print(check_once)
print(estimate)
```
